```python
# Logistic regression model
# Example DataFrame (replace with your actual data)
df = pd.DataFrame({
    'feature1': np.random.randn(100),
    'feature2': np.random.randn(100),
    'target': np.random.randint(0, 2, 100)
})

X = df[['feature1', 'feature2']]
y = df['target']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Add constant term for intercept
X_train = sm.add_constant(X_train)

# Define and fit the model
logm1 = sm.GLM(y_train, X_train, family=sm.families.Binomial())
result = logm1.fit()

# Print the summary
print(result.summary())
```

```python
# Initialize logistic regression model
logreg = LogisticRegression()

# Initialize RFE with logistic regression and the number of features to select
rfe = RFE(estimator=logreg, n_features_to_select=13)

# Fit RFE
rfe = rfe.fit(X_train, y_train)

# Print the selected features
print("Selected features: ", rfe.support_)
print("Ranking of features: ", rfe.ranking_)
```
